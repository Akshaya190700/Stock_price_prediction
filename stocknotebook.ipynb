```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Activation,Dropout
from keras.layers import Dense
from keras.layers import LSTM
from keras.models import Sequential



```

```python

xtrain=pd.DataFrame(df['Close'][0:int(len(df)*0.80)])
xtest=pd.DataFrame(df['Close'][int(len(df)*0.80):int(len(df))])

```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
scalar=MinMaxScaler(feature_range=(0,1))
scaled_data=scalar.fit_transform(xtrain)
```

```python
scaled_data
```

```python
scaled_data.shape[0]
```

```python
X_train=[]
y_train=[]
for i in range(100,scaled_data.shape[0]):
    X_train.append(scaled_data[i-100:i])
    y_train.append(scaled_data[i,0])
X_train,y_train=np.array(X_train),np.array(y_train)
```

```python
X_train.shape
```

```python
model = Sequential()

# First LSTM layer with the correct input_shape
model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape= (X_train.shape[1],1)))
model.add(Dropout(0.2))

# Second LSTM layer
model.add(LSTM(units=60, activation='relu', return_sequences=True))
model.add(Dropout(0.3))

# Third LSTM layer
model.add(LSTM(units=80, activation='relu', return_sequences=True))
model.add(Dropout(0.4))

# Fourth LSTM layer
model.add(LSTM(units=120, activation='relu'))
model.add(Dropout(0.5))

# Output layer
model.add(Dense(units=1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(X_train,y_train,epochs=20,batch_size=32)

```

```python
past_100=xtrain.tail(100)
```

```python
final_data = pd.concat([past_100, xtrain], ignore_index=True)
```

```python
final_data.head()
```

```python
inputscaled_data=scalar.fit_transform(final_data)
```

```python
X_test=[]
y_test=[]
for i in range(100,inputscaled_data.shape[0]):
    X_test.append(inputscaled_data[i-100:i])
    y_test.append(inputscaled_data[i,0])
X_test,y_test=np.array(X_test),np.array(y_test)
```

```python
X_test.shape
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred.shape
```

```python
scalar.scale_
```

```python
scalar_factor=1/0.00537953
y_pred=y_pred*scalar_factor
y_test=y_test*scalar_factor
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Orginal Price', color='green',linewidth=1)
plt.plot(y_pred, label='Predicted Price', color='red',linewidth=1)
plt.legend()
plt.grid()
plt.show()
```
